```python
##Anscombe's Quartet
X1 = [10,8,13,9,11,14,6,4,12,7,5]
X2 = [10,8,13,9,11,14,6,4,12,7,5]
X3 = [10,8,13,9,11,14,6,4,12,7,5]
X4 = [8,8,8,8,8,8,8,19,8,8,8]

Y1 = [8.04,6.95,7.58,8.81,8.33,9.96,7.24,4.26,10.84,4.82,5.68]
Y2 = [9.14,8.14,8.74,8.77,9.26,8.1,6.13,3.1,9.13,7.26,4.74]
Y3 = [7.46,6.77,12.74,7.11,7.81,8.84,6.08,5.39,8.15,6.42,5.73]
Y4 = [6.58,5.76,7.71,8.84,8.47,7.04,5.25,12.5,5.56,7.91,6.89]
```

```python
from matplotlib import pyplot as plt
from sklearn import linear_model
import numpy as np
import math
linear_m = linear_model.LinearRegression()
X1 = np.array(X1)
Y1 = np.array(Y1)
linear_m.fit(X1.reshape(-1,1),Y1)
a,b = linear_m.coef_,linear_m.intercept_
#xmin = math.floor(min(X1))-1
#xmax = math.ceil(max(X1))+1
#ymin = math.floor(min(Y1))-1
#ymax = math.ceil(max(Y1))+1
xmin = 2
ymin = 2
xmax = 20
ymax = 14
plt.axis([xmin,xmax,ymin,ymax])
plt.scatter(X1,Y1,marker = 'o',color = '#3778bf',edgecolors = '#042e60',s = 60,label = 'Position')
plt.plot([xmin,xmax],[a*xmin+b,a*xmax+b],color = '#fa4224',lw = 2,label = 'Curve')
plt.legend(loc = 'upper left')
plt.savefig('F1.png',dpi=600)
```
